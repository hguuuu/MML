# Pass data through LSTM to get feature vectors of size 35
Note that there are 35 action units extracted from each frame. Since each utterance video has different length, the number of frames contained in each video is different, causing variance length of extracted feature vector. We are processing the features through LSTM for one pass to get a uniform feature vector size.


## Sample full_features_dict keys
(['812_14', '47_4', '32_12', '913_1', '678_0', '377_1', '964_3', '251_2', '596_8', '49_19', '369_0', '344_0', '67_3', '604_4', '272_1', '11_0', '463_8', '747_11', '559_5', '448_3', '477_4', '770_3', '40_2', '656_0', '714_4', '640_5', '199_2', '947_7', '605_2', '795_2', '666_3', '507_1', '452_3', '331_9', '558_1', '533_0', '594_5', '666_7', '238_1', '769_1', ...


```python
LSTM_model = nn.LSTM(35, 35)
# LSTM_model.cuda()
```

```python
train_vis_features = {}
h0 = torch.autograd.Variable(torch.ones(35))
c0 = torch.autograd.Variable(torch.ones(35))
```

```python
# # If cuda is available
# for key in tqdm(train_full_vis_features):
#     cur = torch.tensor(list(train_full_vis_features[key].values())).T.to(device="cuda")
#     x, y = cur.shape
#     _, (hn, _) = LSTM_model(cur.reshape(x, 1, y))

#     train_vis_features[key] = hn.reshape(35).cpu().data.numpy()
    
for key in tqdm(train_full_vis_features):
    cur = torch.tensor(list(train_full_vis_features[key].values())).T
    x, y = cur.shape
    _, (hn, _) = LSTM_model(cur.reshape(x, 1, y))

    train_vis_features[key] = hn.reshape(35).data.numpy()
```

# Take 30 time intervals and average the aus features for each.

```python
# # Load full dictionary
# load_data_path = "MML/vis_features_openface/"
# split = 'dev'
# with open(load_data_path + split+'_vis_full_aus.pickle', 'rb') as handle:
#     train_full_vis_features = pickle.load(handle)
```

```python
# len(train_full_vis_features)
```

```python
interval_30_vis_features = {}
for key in train_full_vis_features:
    cur_len = len(train_full_vis_features[key]['AU01_r'])
    cur_val = np.array(list(train_full_vis_features[key].values()))
    if cur_len > 30:
        interval_index = np.append(np.round(np.arange(0, cur_len, cur_len/30))[:30], [cur_len])
        
        cur_new_val = np.mean(cur_val[:, int(interval_index[0]): int(interval_index[1])], axis = 1).reshape(35, 1)
#         print(cur_val.shape, cur_new_val.shape)
#         print(interval_index, cur_val[0])
        for i in range(1, 30):
#             print(cur_new_val)
            cur_new_val = np.hstack([cur_new_val, np.mean(cur_val[:, int(interval_index[i]): int(interval_index[i+1])], axis = 1).reshape(35, 1)])
        interval_30_vis_features[key] = cur_new_val
#         print(interval_index, cur_val[0], interval_30_vis_features[key])
    else:
        interval_30_vis_features[key] = cur_val
```

```python
with open(split+'_vis_30_intervals_aus.pickle', 'wb') as handle:
    pickle.dump(interval_30_vis_features, handle)
```

```python
frame_30_vis_features = {}
for key in train_full_vis_features:
    cur_len = len(train_full_vis_features[key]['AU01_r'])
    cur_val = np.array(list(train_full_vis_features[key].values()))
    if cur_len > 30:
        interval_index = np.round(np.arange(0, cur_len, cur_len/30))[:30].astype(int)
#         print(interval_index)
#         print(np.mean(cur_val, axis=1).reshape(35, 1), np.std(cur_val, axis=1).reshape(35, 1), np.max(cur_val, axis=1).reshape(35, 1))
#         print(np.concatenate([cur_val[:, interval_index], np.mean(cur_val, axis=1).reshape(35, 1), np.std(cur_val, axis=1).reshape(35, 1), np.max(cur_val, axis=1).reshape(35, 1)], axis=1))
#         cur_new_val = .reshape(35, 1)
        frame_30_vis_features[key] = np.concatenate([cur_val[:, interval_index], np.mean(cur_val, axis=1).reshape(35, 1), np.std(cur_val, axis=1).reshape(35, 1), np.max(cur_val, axis=1).reshape(35, 1)], axis=1)
#         print(frame_30_vis_features[key].shape)
    else:
        frame_30_vis_features[key] = cur_val
```
